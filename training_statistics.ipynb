# Training statistics

This notebook should show how models were trained and their accuracies on training as well as on validation set.

```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib
%matplotlib inline
```

## Preprocess data

```python
input_path = pathlib.Path("./trained_models/")
training_accuracy = {}
validation_accuracy = {}
for folder in list(input_path.glob("*")):
    if (folder/"traning_metrics.csv").exists():
        df = pd.read_csv(folder/"traning_metrics.csv", index_col=0)
        training_accuracy[folder.name]= df["accuracy"]
        validation_accuracy[folder.name]= df["val_accuracy"]
```

## Training accuracies

```python
training_df = pd.DataFrame.from_dict(training_accuracy)*100
training_df.plot(figsize = (18,10),title = "Accuracy on train dataset", xlabel="Epochs", ylabel="Accuracy in %")
```

Here we can see maximal accuracies of all networks on training dataset

```python
training_df.max()
```

## Validation accuracies

```python
validation_df = pd.DataFrame.from_dict(validation_accuracy)*100
validation_df.plot(figsize = (18,10),title = "Accuracy on val dataset", xlabel="Epochs", ylabel="Accuracy in %")
```

Here we can see maximal accuracies of all networks on validation dataset

```python
validation_df.max()
```
